## For anidado (nested for)

```python
for f in m2: # For más externo
    for c in f: # For más interno
        print(c, end=' ')
    print()

"""
(0, 0) (0, 1) (0, 2)
(1, 0) (1, 1) (1, 2)
(2, 0) (2, 1) (2, 2)
"""
```

```python
for f in range(len(m2)): # For más externo
    for c in range(len(m2[0])): # For más interno
        print('(', f, ',', c, ')', end=' ', sep='')
    print()

"""
(0, 0) (0, 1) (0, 2)
(1, 0) (1, 1) (1, 2)
(2, 0) (2, 1) (2, 2)
"""
```

```python
for f in m2: # For más externo
    for c in f: # For más interno
        print(c * 10, end=' ')
    print()
```

# random

```python
import random
```

```python
random.random()
```

```python
lista_alumnos = ['Fernanda', 'Diego', 'Andrea', 'Ángel', 'Marco', 'Óscar']

random.choice(lista_alumnos)
```

```python
random.sample(lista_alumnos, 2)
```

```python
# Una función que al llamarla simule tirar un dado [1, 6]

# def dado():
#     cara = random.randint(1, 6)
#     print(cara)

def tirarDado():
    print(f'Tiraste el dado y cayó: {random.randint(1, 6)}')

tirarDado()
```

```python
# Una función que al llamarla simule lanzar una moneda -> águila o sol
# random.random() > .5 : águila < .5 : sol

def lanzarMoneda():
    res = random.random()
    if res > .5:
        print('Águila')
    else:
        print('Sol')

lanzarMoneda()
```

```python
# Con el ciclo que quieran: llenar una lista con 100 valores enteros aleatorios [0, 100]
lista = []

for x in range(100):
    lista.append(random.randint(0, 100))

print(lista)
```

```python
lista = [random.randint(0, 100) for _ in range(100)]
print(lista)
```

```python
# Generar una matriz (lista de listas) de 5x5 con enteros aleatorios [1, 50]

m5x5 = []
for i in range(5):
    tmp = []
    for j in range(5):
        tmp.append(random.randint(1, 50))
    m5x5.append(tmp)

m5x5
```

```python
m5x5 = [[random.randint(1, 50) for i in range(5)] for j in range(5)]
print(m5x5)
```
