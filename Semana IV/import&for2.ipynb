# Emulando matrices

```python
# Matriz 2x2 (2 sublistas, 2 elementos en c/u)
# mnemónico: nombre que hace alución a lo que guarda (que es fácil de recordar)
# fi-co (filas, columnas)

m1 = [
   [1, 2], 
   [3, 4]
]

m1[1][0]
```

```python
# Matriz 3x3
m2 = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

type(m2)
print(m2[1][1])
```

## For anidado (nested for)

```python
for f in m2: # For más externo
    for c in f: # For más interno
        print(c, end=' ')
    print()

"""
(0, 0) (0, 1) (0, 2)
(1, 0) (1, 1) (1, 2)
(2, 0) (2, 1) (2, 2)
"""
```

```python
for f in range(len(m2)): # For más externo
    for c in range(len(m2[0])): # For más interno
        print('(', f, ',', c, ')', end=' ', sep='')
    print()

"""
(0, 0) (0, 1) (0, 2)
(1, 0) (1, 1) (1, 2)
(2, 0) (2, 1) (2, 2)
"""
```

```python
for f in m2: # For más externo
    for c in f: # For más interno
        print(c * 10, end=' ')
    print()
```
