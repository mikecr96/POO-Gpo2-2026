```python
verdadero = True
booleanos = [True, True, True]

for i in range(len(booleanos)):
    if bool(i) == False:
        verdadero = False

print(verdadero)
```

```python
contador = 0
booleanos = [True, True, True]
for j in booleanos:
    if j == True:
        contador += 1
if contador == len(booleanos):
    print(True)
else:
    print(False)
```

```python
# Una función que pida 3 números por teclado y los sume (A+B+C)

def sumar():
    a = int(input('Ingrese el primer valor'))
    b = int(input('Ingrese el segundo valor'))
    c = int(input('Ingrese el tercer valor'))
    print(f'El resultado de sumar {a} + {b} + {c} es: {a + b + c}')

sumar()
```

```python
# Una función que pida 3 números por teclado y los reste (A-B-C)

def restar():
    a = int(input('Ingrese el primer valor'))
    b = int(input('Ingrese el segundo valor'))
    c = int(input('Ingrese el tercer valor'))
    print(f'El resultado de restar {a} - {b} - {c} es: {a - b - c}')

restar()
```
