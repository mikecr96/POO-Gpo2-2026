```python
from datetime import datetime, timezone, timedelta

# datetime.now()
# print(timezone.utc)

# 1. Crear una fecha en UTC
dt_utc = datetime.now(timezone.utc)
print("UTC:", dt_utc)

# # 2. Convertir a una zona horaria específica (ej. Madrid/UTC+1)
# timezone_madrid = timezone(timedelta(hours=1))
# dt_madrid = dt_utc.astimezone(timezone_madrid)
# print("Madrid:", dt_madrid)

# 3. Convertir a otra zona horaria (ej. UTC-6)
timezone_mexico = timezone(timedelta(hours=-6))
dt_mexico = dt_utc.astimezone(timezone_mexico)
print("CDMX:", dt_mexico)
```

```python
# clases, objetos, atributos, propiedades, herencia, polimorfismo, etc...

class Animal:
    # Método iniciador (en otros lenguajes constructor)
    def __init__(self, nombre, edad, especie, ruido):
        self.nombre = nombre
        self.edad = edad
        self.especie = especie
        self.ruido = ruido
    # Cuando una función está dentro de una clase se llama Método
    def hacer_ruido(self):
        print(f"{self.nombre} hizo {self.ruido}")
```

```python
# Instanciar una clase
melman = Animal('Melman', 10, 'jirafa', 'ruido_en_jirafa') # OBJETO
marty = Animal('Marty', 50000, 'zebra', 'rebusna')
skipper = Animal('Skipper', 3, 'pingüino', 'grasna')
# Notación punto (dot-notation)
skipper.hacer_ruido()
```
