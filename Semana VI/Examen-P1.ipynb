### ü™ô Fase 2: El Oro del Sin Cara

**Ejercicio 3.** El esp√≠ritu "Sin Cara" est√° arrojando pepitas de oro. Crea una lista vac√≠a. Utiliza un ciclo `while` para solicitar al usuario que ingrese los distintos pesos del oro recolectado (n√∫meros que pueden contener decimales). Agrega cada peso a la lista. El ciclo debe detenerse exclusivamente cuando el usuario ingrese el n√∫mero `0`. **Nota importante:** el n√∫mero cero que rompe el ciclo *no* debe agregarse a tu lista.

**Ejercicio 4.** A partir de la lista generada en el Ejercicio 3, imprime en la salida de tu celda la siguiente informaci√≥n: la cantidad total de pepitas recolectadas, la suma total del peso del oro, la pepita m√°s pesada y la pepita m√°s ligera. *(Pista: utiliza las funciones built-in de Python para obtener estos valores).*

### üõÅ Fase 3: Limpiando al Dios del R√≠o

**Ejercicio 5.** El Dios del R√≠o ha llegado a la ba√±era principal y ha dejado mucho lodo. Escribe el c√≥digo necesario para generar una matriz de 100x100 (es decir, una lista que contenga 100 listas, y cada una de ellas con 100 elementos) conformada por n√∫meros enteros aleatorios entre el `-500` y el `500` que representan la densidad del lodo.

**Ejercicio 6.** Utiliza ciclos `for` anidados para recorrer la matriz de 100x100 del ejercicio anterior y realiza lo siguiente:
* Identifica todos los n√∫meros que sean **exactamente iguales a `0`** (estos representan amuletos m√°gicos ocultos en el lodo).
* Cada vez que encuentres uno, imprime un mensaje que indique su posici√≥n (√≠ndices de la fila y la columna). Formato esperado: *"Amuleto encontrado en las coordenadas ([Fila], [Columna])"*.
* Lleva un registro y, al finalizar todo el recorrido de la matriz, imprime la cantidad total de amuletos encontrados.

```python
import random

matriz = []
for i in range(100):
    matriz.append([random.randint(-500, 500)])
    for j in range(100):
        matriz[i].append(random.randint(-500, 500))

amuletos_encontrados = 0
for k in range(100):
    for j in range(100):
        if matriz[k][j] == 0:
            print(f"Amuleto encontrado en las coordenadas: [{k}, {j}]")
            amuletos_encontrados += 1

print(f"La cantidad total de amuletos encontrados fue: {amuletos_encontrados}")
```

**Ejercicio 7.** Genera una lista llamada `energia_espiritus` que contenga 40 n√∫meros enteros aleatorios entre `100` y `1000`, representando el nivel de energ√≠a de los pasajeros que esperan el tren.

**Ejercicio 8.** Utilizando *List Comprehension* en una sola l√≠nea de c√≥digo, crea una nueva lista llamada `espiritus_debiles` que filtre y guarde √∫nicamente los valores de la lista del ejercicio anterior que sean estrictamente menores a `400`.

**Ejercicio 9.** Recorre la lista `espiritus_debiles` y utiliza **f-strings** para imprimir cada uno de sus valores siguiendo exactamente este formato: 
*"Aviso a Kamaji: Esp√≠ritu d√©bil detectado con energ√≠a de -> [valor]"*

```python
energia_espiritus = [random.randint(100, 1000) for _ in range(40)]

espiritus_debiles = [energia for energia in energia_espiritus if energia < 400]

for energia in espiritus_debiles:
    print(f"Aviso a Kamaji: Esp√≠ritu d√©bil detectado con enerfia de -> {energia}")
```
