```python
# EN PYTHON NO EXISTE EL TIPO DE DATO 'CHAR'. SOLO EXISTE EL STR
# PEP

cadena_vacia = ""
print(type(cadena_vacia))

cadena_uncaracter1 = 'm'
print(type(cadena_uncaracter1))

cadena_uncaracter2 = "m"
print(type(cadena_uncaracter2))
```

```python
# Booleanos (sí o no)

verdadero = True
falso = False
print(type(verdadero))
print(type(falso))
```

```python
# list
lista = []
type(lista)
```

```python
lista = [123, 1231, 12312, 1231, 12.3, True, []]
type(lista)
```

```python
# tuple: lista inmutable
abc = ()
print(type(abc))

cba = (1, 2, True, [], ())
print(type(cba))

# tupla de un solo caracter
tdsc = ("1",)
print(type(tdsc))
```

```python
# set (No aceptan repetición)
conjunto = {1, 1, 1, 1, 1, 1, 2, 3, 4}
conjunto
```

```python
# diccionarios (dict)
# juego {llave : valor}
diccionario = {"gato" : "mamífero perteneciente a los felinos", "perro" : "Mamífero."}
type(diccionario)
```
