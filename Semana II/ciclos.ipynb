```python
i = 0

while i < 10:
    print(i, "Hola mundo")
    i += 1 # i = i + 1
```

```python
i = 0

while i < 1000000:
    print(i, "Hola mundo")
    i += 1 # i = i + 1
    if i == 100:
        break
```

## FOR

```python
# for(int i=0; i < 100; i++)

lista = [1, 2, 3]

for j in lista:
    print(j)
```

```python
string = "Hola alumnos."

for s in string:
    print(s)
```

```python
# for(int i=0; i < 100; i++)
# n = 123
for numero in range(10, 100, 2):
    print(numero)
```

```python
lista = []

l = 0

while l < 100:
    l += 1
    lista.append(l)

print(lista)
```

```python
lista = []

for contador in range(0, 101, 1):
    lista.append(contador)

print(lista)
```
