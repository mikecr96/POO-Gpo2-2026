# IF - ELSE

```python
if 3 > 2:
    print("Hola mundo")
```

```python
if 3 < 2: # La condición no se cumplió
    print("Hola mundo")
```

```python
booleano = 1 != 1

if booleano:
    print("SÍ se cumplió la condición")
else:
    print("NO se cumplió la condición")
```

```python
numero = "Hola mundo 100!"

if numero.isdigit():
    # numero = int(numero)
    if int(numero) % 2 == 0:
        print("Tu número es par")
    else:
        print("Tu número es impar")
else:
    print("TU string no está formado solo por digitos")
```
