```python
"""
Usuario ingrese día, mes, año (input). La convierto a fecha.
Imprimo que día de la semana es (lun, mar, etc...) y que diga si es finde o no (vie-dom)
"""
```

```python
dia = input("Ingrese día:")
mes = input("Ingrese mes:")
year = input("Ingrese año:")
fecha = f"{dia}/{mes}/{year}"
fecha_chida = datetime.strptime(fecha, "%d/%m/%Y")
dias = ["lun", "mar", "mier", "jue", "vie", "sab", "dom"]

weekday = fecha_chida.weekday()
print(f"Hoy es {dias[weekday]}")
if weekday < 5:
    print("Es entre semana")
else:
    print("Es finde")
```
