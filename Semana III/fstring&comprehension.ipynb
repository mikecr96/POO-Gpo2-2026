# List comprehension

```python
# Llenar la lista con elementos del 1 al 100
# Esto está bien
lista = []

for valor in range(1, 101, 1):
    lista.append(valor)

print(lista)
```

```python
# Llenar la lista con elementos del 1 al 100
# Esto está bien

lista = [i for i in range(1, 101, 1)]

print(lista)
```

```python
# Llenar la lista con elementos del 1 al 100
# Esto está bien

tupla = (i for i in range(1, 101, 1))

tupla
```

```python
# Llenar la lista con elementos del 1 al 100
# Esto está bien

conjunto = {i for i in range(1, 101, 1)}

print(conjunto)
```

```python
# Llenar la lista con elementos del 1 al 100
# Esto está bien

conjunto = {i : i for i in range(1, 101, 1)}

print(conjunto)
```
