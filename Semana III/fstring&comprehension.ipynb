# List comprehension

```python
# Llenar la lista con elementos del 1 al 100
# Esto está bien
lista = []

for valor in range(1, 101, 1):
    lista.append(valor)

print(lista)
```

```python
# Llenar la lista con elementos del 1 al 100
# Esto está bien

lista = [i for i in range(1, 101, 1)]

print(lista)
```

```python
# Llenar la lista con elementos del 1 al 100 elevados al cuadrado

lista = [i ** 2 for i in range(1, 101, 1)]

print(lista)
```

```python
# Llene una lista con valores del 1 al 100, pero solo dejando los pares

lista = []

for i in range(1, 101, 1):
    if i % 2 == 0:
        lista.append(i)

lista
```

```python
# Llene una lista con valores del 1 al 100, pero solo dejando los pares

# lista = [i for i in range(1, 101, 1) if i % 2 == 0 else "*"] # No se puede poner else

lista
```

```python
# Llenar la lista con elementos del 1 al 100
# Esto está bien

tupla = (i for i in range(1, 101, 1))

tupla
```

```python
# Llenar la lista con elementos del 1 al 100
# Esto está bien

conjunto = {i for i in range(1, 101, 1)}

print(conjunto)
```

```python
# Llenar la lista con elementos del 1 al 100
# Esto está bien

diccionario = {i : i for i in range(1, 101, 1)}

print(diccionario)
```

```python
# con dict-comprehension llenar el diccionario donde cada llave es un número del 1 al 100
# y cada valor es la llave elevada al cuadrado

diccionario = {i : i ** 2 for i in range(1, 101, 1)}

print(diccionario)
```

```python
# con dict-comprehension llenar el diccionario donde cada llave es el nombre
# y cada valor es el tamaño (no. de caracteres) de la llave
nombres = ["Miguel", "Ximena", "Leonardo", "Max"]
for nombre in nombres:
    print(nombre)
    print(len(nombre))
```

```python
diccionario = {nombre : len(nombre) for nombre in nombres}
diccionario
```
