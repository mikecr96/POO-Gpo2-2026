```python
# try:
#   {bloque}
# except <Error>:
#   {bloque}
# 10 / 0
```

```python
try:
    print("Hola mundo")
except:
    print("ERROR!")
```

```python
try:
    10 / 0
except:
    print("ERROR!")
```

```python
diccionario = {1: "hola"}
try:
    diccionario[2]
except:
    print("ERROR!")
```

```python
diccionario = {1: "hola"}
try:
    diccionario[2]
except:
    print("ERROR!")
finally: # finally SIEMPRE se ejecuta
    print("Hola mundo")
```

```python
diccionario = {1: "hola"}
try:
    print(diccionario[1])
except:
    print("ERROR!")
finally: # finally SIEMPRE se ejecuta
    print("Hola mundo")
```

```python
try:
    print(diccionario[1])
# except KeyError:
#     print("Error, la llave no existe!")
except ValueError:
    print("Valor incorrecto")
except ZeroDivisionError:
    print("Error, dividiste entre 0")
finally: # finally SIEMPRE se ejecuta
    print("Hola mundo")
```

```python
try:
    # print(diccionario['perro'])
    10 / 0
except KeyError as e:
    print(f"Error, la llave no existe! | {e}")
except ValueError:
    print("Valor incorrecto")
except ZeroDivisionError as e:
    print(f"Error, dividiste entre 0 : {e}")
else: # se entra cuando no entramos a ninguno de los except
    print("Ninguna de las Excepciones se cachó")
finally: # finally SIEMPRE se ejecuta
    print("Hola mundo")
```
