# Diccionarios

llave : valor

```python
type({1, 2, 3, 4})
```

```python
diccionario = {1 : "Miguel", 
               2 : "Diana", 
               3 : "Fernanda"}
```

```python
len(diccionario)
```

```python
# En diccionarios: me devuleve el valor asociado a la llave (si esta existe)
# diccionario[0]

diccionario[2]
```

```python
# Por defecto, iterar en un dict. devuelve las llaves
for d in diccionario:
    print(d)
```

```python
dir(diccionario)
```

```python
for llave in diccionario.keys():
    print(llave)
```

```python
for valor in diccionario.values():
    print(valor)
```

```python
diccionario.items()
```

```python
for llave, valor in diccionario.items():
    print("Llave:", llave, "valor:", valor)
```

# Conjuntos

```python
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}
```

```python
dir(a)
```

```python
# a unión b
a.union(b)
```

```python
# a unión b
# | o.a. para unión entre conjuntos
a | b
```

```python
# a intersección b
a.intersection(b)
```

```python
# a intersección b
# & o.a. para intersección entre conjuntos
a & b
```

```python
print(a)
print(b)
```

```python
a.difference(b)
```

```python
a - b
```

```python
a.symmetric_difference(b)
```
